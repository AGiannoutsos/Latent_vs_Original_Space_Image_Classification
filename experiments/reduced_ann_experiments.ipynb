# LSH experiments

```
import numpy as np 
import matplotlib.pyplot as plt

w = [5000, 10000, 25000, 40000, 50000]
Cratio = [0.056028, 0.902032, 1.06027, 1.02146, 1.00063]
Tratio = [0.00323188, 0.00431672, 0.0451374, 0.209865, 0.850346]


fig, ax = plt.subplots(1,1,  figsize=(10, 5))
ax.plot(w, Cratio, label="Dist LSH / Dist True")
ax.plot(w, Tratio, label="Time LSH / Time True")
ax.legend()
ax.set_title("LSH Ratios")
ax.set_xlabel("W")
ax.set_ylabel("Ratio")
plt.savefig('lsh.png')
pass
```

# Latent space experiments

```
dimension = [10, 20, 50, 100, 200]
Cratio = [1.36443, 1.20611, 1.24237, 1.19265, 1.17869]
Tratio = [0.2044, 0.236088, 0.253484, 0.332256, 0.437622]


fig, ax = plt.subplots(1,1,  figsize=(10, 5))
ax.plot(dimension, Cratio, label="Dist Reduced / Dist True")
ax.plot(dimension, Tratio, label="Time Reduced / Time True")
ax.legend()
ax.set_title("Reduced Ratios")
ax.set_xlabel("dimension")
ax.set_ylabel("Ratio")
plt.savefig('reduced.png')
pass
```

# Results

The LSH from the experiments seems to have very good results when the number of w is 20000. The calculation time remains short and the accuracy is quite high.

The reduced dimension has a very short calculation time when the dimensions are small. But accuracy is not close to LSH results. As we increase the dimension, the accuracy increases as the calculation time increases. However, the computation time grows faster than the accuracy, therefore a small number of dimensions close to 20.
